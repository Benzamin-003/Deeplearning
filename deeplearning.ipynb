```python

import math
import matplotlib.pyplot as plt


x = [[1], [2], [3], [4], [5]]
y = [0, 0, 1, 1, 1]

N = len(x)
w = 0
b = 0
epochs = 1000

for i in range(epochs):
    total_error = 0

    for j in range(N):
        z = w * x[j][0] + b
        y_pred = int(1 / (1 + math.exp(-z)) >= 0.6)
        error = y[j] - y_pred

        w = w + x[j][0] * error
        b = b + error

        total_error += error ** 2

    if i % 100 == 0:
        mse = total_error / N
        print(f"Epoch {i} MSE = {mse:.4f}, w = {w}, b = {b}")

print(f"\nFinal weights: w = {w}, b = {b}")
```

```python

x_vals = [x_i[0] for x_i in x]
y_vals = y

plt.scatter(x_vals, y_vals, c=y_vals, cmap='bwr', label='Data Points')

boundary_x = [min(x_vals), max(x_vals)]
boundary_y = [1 / (1 + math.exp(-(w * x_val + b))) for x_val in boundary_x]
plt.plot(boundary_x, boundary_y, label='Decision Boundary')

plt.title('Logistic Regression: Data Points and Decision Boundary')
plt.xlabel('X')
plt.ylabel('Predicted Probability')
plt.grid(True)
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
import numpy as np


model = LogisticRegression(max_iter=1000, solver='liblinear')
model.fit(x, y)

y_pred = model.predict(x)

mse = mean_squared_error(y, y_pred)

print(f"Scikit-learn Logistic Regression:")
print(f"Coefficients: {model.coef_}, Intercept: {model.intercept_}")
print(f"MSE: {mse:.4f}")
```

```python
plt.scatter(x, y, c=y, cmap='bwr', label='Data points')

boundary_x = np.array([min(x), max(x)])
boundary_y = model.coef_[0] * boundary_x + model.intercept_


plt.plot(boundary_x, boundary_y, color='green', label='Decision Boundary')

plt.title('Logistic Regression from Scikit-learn')
plt.xlabel('X')
plt.ylabel('Predicted Probability')
plt.legend()
plt.grid(True)
plt.show()
```
